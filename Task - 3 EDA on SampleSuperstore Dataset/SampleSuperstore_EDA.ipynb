### Grouping Sales, Quantity and Profit Category and Sub-Category wise

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,20))
sns.catplot(x="Profit", y="Sub-Category", col="Category", data=category_wise, kind='bar')
```

**Observations:**

From the above graphs we can conclude the following: 

* Under the furniture category, Tables and Bookcases have been incurring losses.
* In Office Supplies category, theres minimal profit amongst the items.
* In technology category, Copiers have been the most profitable.

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,20))
sns.catplot(x="Quantity", y="Sub-Category", col="Category", data=category_wise, kind='bar')
```

### Grouping Sales, Quantity and Profit Segment wise

```python
segment_wise = data.groupby(['Segment'])['Sales','Quantity','Profit'].mean().reset_index()
segment_wise
```

```python

sns.set_style("whitegrid")
plt.figure(figsize = (10,8))
ax = sns.catplot(x='Segment', y = 'Sales', data = segment_wise, palette = 'Set1', kind = 'bar')

for ax in ax.axes.ravel():
  for p in ax.patches:
    ax.annotate(format(p.get_height(),'.2f'), xy=(p.get_x() + p.get_width() / 2,  p.get_height()), ha='center',  va='center',size=12, xytext=(0, 8),textcoords='offset points')

plt.title("Segment wise Sales",fontsize = 12)
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize = (10,8))
ax = sns.catplot(x='Segment', y = 'Profit', data = segment_wise, palette = 'Set1', kind = 'bar')

for ax in ax.axes.ravel():
  for p in ax.patches:
    ax.annotate(format(p.get_height(),'.2f'), xy=(p.get_x() + p.get_width() / 2,  p.get_height()), ha='center',  va='center',size=12, xytext=(0, 8),textcoords='offset points')

plt.title("Segment wise Profit made",fontsize = 12)
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize = (10,8))
ax = sns.catplot(x='Segment', y = 'Quantity', data = segment_wise, palette = 'Set1', kind = 'bar')

for ax in ax.axes.ravel():
  for p in ax.patches:
    ax.annotate(format(p.get_height(),'.2f'), xy=(p.get_x() + p.get_width() / 2,  p.get_height()), ha='center',  va='center',size=12, xytext=(0, 8),textcoords='offset points')

plt.title("Segment wise quantity sold",fontsize = 12)
plt.show()
```

**Observations:**

From the above graphs we can conclude that:


*   All Segments have sold almost same number of quantities.
*   All Segments have done almost similar Sales.
* Home Office Segment has made the most profit.



### Grouping Sales, Quantity and Profit State wise

```python
state_wise = data.groupby(['State'])['Sales','Quantity','Profit'].mean().reset_index()
```

```python
sorted_sales = state_wise.sort_values(by = ['Sales'], ascending = False)
States = sorted_sales['State']
Sales = sorted_sales['Sales']
fig = plt.figure(figsize =(30, 10))

sns.barplot(States, Sales, palette = 'rocket')
plt.xticks(rotation=45)
plt.show()
```

```python
sorted_profit = state_wise.sort_values(by = ['Profit'], ascending = False)
States = sorted_profit['State']
Profit = sorted_profit['Profit']
fig = plt.figure(figsize =(30, 10))

sns.barplot(States, Profit, palette = 'Blues_r')
plt.xticks(rotation=45)
plt.show()
```

```python
sorted_quantity = state_wise.sort_values(by = ['Quantity'], ascending = False)
States = sorted_quantity['State']
Quantity = sorted_quantity['Quantity']
fig = plt.figure(figsize =(30, 10))

sns.barplot(States, Quantity, palette = 'Greens_r')
plt.xticks(rotation=45)
plt.show()
```

**Observations:**

From the above plots we can infer that:
* Vermont state has the maximum profit yielding business whereas Ohio has the maximum loss making business.
* Wyoming has the maximum sales and South Dakota has the minimum sales.
